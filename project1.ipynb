```python
import random
```

```python
ai=random.choice([-1,0,1])
youdict={"s":-1,"w":0,"g":1}
revdict={-1:"snake",0:"water",1:"gun"}
youstr=input("choose s,w,g:")
```

```python
you=youdict[youstr]
```

```python
print(f"You choose {revdict[you]}\nComputer chose {revdict[ai]}")

if(ai == you):
  print("Its a draw")
else:
  if(ai==-1 and you==1): print("You win!")
  elif(ai==-1 and you ==0): print("You Lose!")
  elif(ai==1 and you==-1):print("You lose!")
  elif(ai==1 and you==0): print("You Win!")
  elif(ai==0 and you ==-1): print("You Win!")
  elif(ai==0 and you==1): print("You Lose!")
  else:
    print("Something went wrong!")
```
